# Ejercicios de probabilidad

## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
def simulate_dice(tries=1000):
    successes = 0

    for _ in range(tries):
        dice_a = random.randint(1, 6)
        dice_b = random.randint(1, 6)

        total = dice_a + dice_b

        if total > 7 or total % 2 == 0:
            successes += 1

    probability = successes / tries
    return probability

result = simulate_dice()
print(f"Probabilidad estimada: {np.round(result * 100, 2)}%")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
box = {}

for idx in range(60):
    if idx < 10:
        box[idx] = "White"
    elif idx < 30:
        box[idx] = "Red"
    else:
        box[idx] = "Green"


def simulate_draws(trials=1000):
    three_white_two_red = 0
    same_color = 0

    for _ in range(trials):
        drawn_colors = []

        # Draw 5 balls with replacement
        for _ in range(5):
            random_ball = np.random.randint(0, 60)
            drawn_colors.append(box[random_ball])

        drawn_colors = np.array(drawn_colors)

        whites = np.sum(drawn_colors == "White")
        reds = np.sum(drawn_colors == "Red")
        greens = np.sum(drawn_colors == "Green")

        # Case 1: 3 white and 2 red
        if whites == 3 and reds == 2:
            three_white_two_red += 1

        # Case 2: all same color
        if whites == 5 or reds == 5 or greens == 5:
            same_color += 1

    return three_white_two_red / trials, same_color / trials


p1, p2 = simulate_draws(1000)

print(f"Probabilidad de 3 blancas y 2 rojas: {np.round(p1 * 100, 2)}%")
print(f"Probabilidad de todas del mismo color: {np.round(p2 * 100, 2)}%")
```
